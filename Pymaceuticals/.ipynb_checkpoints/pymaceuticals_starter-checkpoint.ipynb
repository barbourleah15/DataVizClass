```python
# Resources: Geekforgeeks.org
# https://towardsdatascience.com/a-guide-to-pandas-and-matplotlib-for-data-exploration-56fad95f951c
# datasciencemadesimple.com
# stackoverflow.com
# https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.plot.html#matplotlib.axes.Axes.plot
# In the Survival Rates graph, the example png given did not include error bars and even though the Readme file said all
# scatter plots had to have them, I excluded them because it didn't make since to include nor did they match the example.
# Similarly, the survival graph y-axis has been modified to reflect percentages though not specified in the Readme.
# https://www.afterhoursprogramming.com/tutorial/python/tuples/

# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "Resources/mouse_drug_data.csv"
clinical_trial_data_to_load = "Resources/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load, low_memory=False)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load, low_memory=False)

# Combine the data into a single dataset
merged_data = mouse_drug_data.merge(clinical_trial_data, on="Mouse ID")

# Display the data table for preview
merged_data
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame

tumor_volume_data = merged_data.loc[:,["Drug", "Timepoint", "Tumor Volume (mm3)"]]
tumor_data_df=tumor_volume_data.groupby(["Drug", "Timepoint"]).mean()
tumor_data_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# Convert to DataFrame

# Preview DataFrame
tumor_sems = tumor_volume_data.groupby(["Drug", "Timepoint"]).sem()
# tumor_sems.head()
tumor_sems
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked
tumor_table = pd.pivot_table(tumor_data_df,  columns = ["Drug"], index = ["Timepoint"])
tumor_table.head(10)
```

```python
tumor_table.max()
```

```python
tumor_table.min()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

# Convert to DataFrame

# Preview DataFrame
```

```python
survival_data = merged_data.loc[:,["Drug", "Timepoint", "Mouse ID"]]
survival_count = survival_data.groupby(["Drug", "Timepoint"]).count()
survival_count.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview the Data Frame
```

```python
survival_count = pd.pivot_table(survival_count, columns = ["Drug"], index = ["Timepoint"])
survival_count.head(10)
```

```python
# Generate the Plot (Accounting for percentages)

# Save the Figure

# Show the Figure
fig2=plt.figure(figsize=(12,12))

x_axis = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]
x_limit = 50

Cap_line = plt.errorbar(x_axis, survival_count["Mouse ID"]["Capomulin"]/25*100, marker = "D", color = "red", label = "Capomulin", ls = "--")

Cef_line = plt.errorbar(x_axis, survival_count["Mouse ID"]["Ceftamin"]/25*100, marker = "D", color = "orange", label = "Ceftamin", ls = "--")

Inf_line = plt.errorbar(x_axis, survival_count["Mouse ID"]["Infubinol"]/25*100, marker = "D", color = "yellow", label = "Infubinol", ls = "--")

Ket_line = plt.errorbar(x_axis, survival_count["Mouse ID"]["Ketapril"]/25*100, marker = "D", color = "green", label = "Ketapril", ls = "--")

Naf_line = plt.errorbar(x_axis, survival_count["Mouse ID"]["Naftisol"]/25*100, marker = "D", color = "blue", label = "Naftisol", ls = "--")

Pla_line = plt.errorbar(x_axis, survival_count["Mouse ID"]["Placebo"]/25*100, marker = "D", color = "purple", label = "Placebo", ls = "--")

Pro_line = plt.errorbar(x_axis, survival_count["Mouse ID"]["Propriva"]/26*100, marker = "D", color = "black", label = "Propriva", ls = "--")

Ram_line = plt.errorbar(x_axis, survival_count["Mouse ID"]["Ramicane"]/25*100, marker = "D", color = "brown", label = "Ramicane", ls = "--")

Ste_line = plt.errorbar(x_axis, survival_count["Mouse ID"]["Stelasyn"]/26*100, marker = "D", color = "grey", label = "Stelasyn", ls = "--")

Zon_line = plt.errorbar(x_axis, survival_count["Mouse ID"]["Zoniferol"]/25*100, marker = "D", color = "pink", label = "Zoniferol", ls = "--")

plt.title("Survival During Treatment", fontsize=16)
plt.xlabel("Time (Days)", fontsize=12)
plt.ylabel("Survival Rates (%)", fontsize=12)

plt.grid(linestyle="dashed")
plt.legend(("Capomulin","Ceftamin",  "Infubinol", "Ketapril", "Naftisol", "Placebo", "Propriva",
           "Ramicane", "Stelasyn", "Zoniferol"), fontsize=10, bbox_to_anchor=(1,1))

plt.show()
```

## Summary Bar Graph

```python
tumor_unstacked = tumor_data_df.unstack(0)
tumor_unstacked
```

```python
# Calculate the percent changes for each drug

# Display the data to confirm
tumor_unstacked = tumor_data_df.unstack(0)
tumor_changes = tumor_unstacked["Tumor Volume (mm3)"]
tumor_changes.head()

change_cap = ((tumor_changes["Capomulin"][45] - tumor_changes["Capomulin"][0])/45)*100
change_cef = ((tumor_changes["Ceftamin"][45] - tumor_changes["Ceftamin"][0])/45)*100
change_inf = ((tumor_changes["Infubinol"][45] - tumor_changes["Infubinol"][0])/45)*100
change_ket = ((tumor_changes["Ketapril"][45] - tumor_changes["Ketapril"][0])/45)*100
change_naf = ((tumor_changes["Naftisol"][45] - tumor_changes["Naftisol"][0])/45)*100
change_pla = ((tumor_changes["Placebo"][45] - tumor_changes["Placebo"][0])/45)*100
change_pro = ((tumor_changes["Propriva"][45] - tumor_changes["Propriva"][0])/45)*100
change_ram = ((tumor_changes["Ramicane"][45] - tumor_changes["Ramicane"][0])/45)*100
change_ste = ((tumor_changes["Stelasyn"][45] - tumor_changes["Stelasyn"][0])/45)*100
change_zon = ((tumor_changes["Zoniferol"][45] - tumor_changes["Zoniferol"][0])/45)*100

change_drug = pd.DataFrame({"Drug": ["Capomulin", "Ceftamin", "Infubinol", "Ketapril", "Naftisol", "Placebo", "Propriva",
                                    "Ramicane", "Stelasyn", "Zoniferol"],
                               "Percentage Change": [change_cap, change_cef, change_inf, change_ket, change_naf, 
                                                    change_pla, change_pro, change_ram, change_ste, change_zon]})

change_drug 
```

```python
change_drug["+/-"] = np.where(change_drug["Percentage Change"]>0.0, "red", "green")
change_drug
```

```python
drugs = ("Capomulin", "Ceftamin", "Infubinol", "Ketapril", "Naftisol", "Placebo", "Propriva",
         "Ramicane", "Stelasyn", "Zoniferol")

fig=plt.figure(figsize=(12,12))

plt.bar(drugs, change_drug["Percentage Change"],width=1, align="center", edgecolor="black")

plt.title("Tumor Change Over 45 Day Treatment", fontsize=16)
plt.xlabel("Drug", fontsize=12)
plt.ylabel("% Tumor Volume Change", fontsize=12)

plt.grid(linestyle="dashed")

#i'm having trouble pulling together the change color, which is correctly being identified, and the graph, 
#which is correctly being graphed.

```

```python
#this is not graphing with the correct colors though the data is still correct. Without quotations on the "0" it returns
#a TypeError: '<' not supported between instances of 'str' and 'int'.
colors = []

for num in change_drug:
    if num < "0":
        colors.append("green")
    else:
        colors.append("red")
        
fig=plt.figure(figsize=(12,12))

plt.title("Tumor Change Over 45 Day Treatment", fontsize=16)
plt.xlabel("Drug", fontsize=12)
plt.ylabel("% Tumor Volume Change", fontsize=12)

plt.grid(linestyle="dashed")

plt.bar(drugs, change_drug["Percentage Change"], color=colors ,width=1, align="center", edgecolor="black")
```
